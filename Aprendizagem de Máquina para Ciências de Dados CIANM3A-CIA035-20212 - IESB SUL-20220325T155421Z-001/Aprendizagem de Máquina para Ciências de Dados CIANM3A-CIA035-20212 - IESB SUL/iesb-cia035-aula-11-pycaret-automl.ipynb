## Problema de Classificação

```python
# Carregando os dados do dataset
df_heart = pd.read_csv('../input/heart-failure-prediction/heart.csv')
df_heart.head()
```

```python
# Analisando os dados
df_heart.info()
```

```python
# Analisando a variável target
df_heart['HeartDisease'].value_counts(normalize=True)
```

```python
# Separando as colunas numéricas e categóricas
# no caso das colunas numéricas precisa excluir a variável target
num = df_heart.drop(['HeartDisease'], axis=1).select_dtypes('number').columns

cat = df_heart.select_dtypes('object').columns

num, cat
```

```python
# Forçando a Series do pandas a virar uma lista
list(cat)
```

```python
# Importando as funções de classificação
from pycaret.classification import *
```

```python
# A função setup deve ser chamada antes de todas, para a preparar o ambiente do PyCaret
# Precisa obrigatoriamente de data e target
setup1 = setup(data = df_heart, 
               target = 'HeartDisease',
               categorical_features = list(cat), 
               silent = True)
```

```python
# Vamos ver os modelos ded classificação disponíveis
models()
```

```python
# Comparando o desempenho dos modelos
top3 = compare_models(sort='Recall',n_select = 3)
```

```python
# Quais os top3 modedlos?
top3
```

```python
# Vamos criar o modelo top1
catboost = create_model('catboost')
```

```python
# Agora vamos melhorar o modelo top1
tuned_catboost = tune_model(catboost, optimize = 'Recall')
```

```python
# Avaliando o modelo melhorado
evaluate_model(tuned_catboost)
```

```python
# Executando as previsões na base de validação
predict_model(tuned_catboost)
```

```python
# Fazendo um blend (uma mistura) dos top3 classificadores
blender_top3 = blend_models(top3, method = 'soft', optimize = 'Recall')
```

```python
# Fazendo um stack dos top3 classificadores
stack_top3 = stack_models(top3, optimize = 'Recall')
```

## Problema de Regressão

```python
# Carregando os dados
df_car = pd.read_csv('../input/car-price-prediction/CarPrice_Assignment.csv')
df_car.head()
```

```python
# Verificando os dados
df_car.info()
```

```python
# Feature Engineering e Correção de Dados

# Separando o modelo
df_car['model'] = [x.split()[0] for x in df_car['CarName']]

# Corrigindo os nomes
df_car['model'] = df_car['model'].replace({'maxda': 'Mazda','mazda': 'Mazda', 
                                     'nissan': 'Nissan', 
                                     'porcshce': 'Porsche','porsche':'Porsche', 
                                     'toyouta': 'Toyota', 'toyota':'Toyota',
                            'vokswagen': 'Volkswagen', 'vw': 'Volkswagen', 'volkswagen':'Volkswagen'})
```

```python
# Importando a bilbioteca
from pycaret.regression import *
```

```python
# Separando as colunas categóricas
cat = df_car.select_dtypes('object').columns

cat
```

```python
# Fazendo o setup do PyCaret
setup2 = setup(data = df_car, 
               target = 'price',
               categorical_features = list(cat) , 
               ignore_features = ['car_ID','CarName'],
               normalize = True,
               silent = True)
```

```python
# Quais os modelos de regressão disponíveis
models()
```

```python
# Comparando os modelos com base em uma lista selecionada
# e escolhendo os top3
top_3 = compare_models(sort='RMSE', include=['lightgbm','catboost','xgboost','rf','mlp','gbr','lasso','ridge','et','ada'], n_select = 3)
```

```python
# Criando um modelo baseado no top1
gbr = create_model('gbr')
```

```python
# Melhorando o modelo top1
tuned_gbr = tune_model(gbr, optimize = 'RMSE')
```

```python
# Avaliando o modelo
evaluate_model(gbr)
```

```python
# Blend de modelos com os top3
blender_top3 = blend_models(top_3, optimize = 'RMSE')
```

```python
# Stack de modelos com os top3
stack_top3 = stack_models(top_3, optimize = 'RMSE')
```

```python
# Criando o modelo específico do lightgbm
lgbm = create_model('lightgbm')
```

```python
# Fazendo tunning com o lightgbm
tuned_lgbm = tune_model(lgbm, optimize = 'RMSE')
```

```python
# Criando o modelo específico para ao RFR
rfr = create_model('rf')
```

```python
# Tunning do RFR
tuned_rfr = tune_model(rfr)
```
