**Questão 01:** Uma moeda viciada tem probabilidade de cara igual a 0.4. Para quatro lançamentos independentes dessa moeda, estude o comportamento da variável
numero de caras e faca um gráfico de sua função de distribuição.

```r
x <- 1:4 
fx <- dbinom(x, 4, 0.4)
plot(x, fx, type ='h', main = "Função de distribuição de probabilidade", lwd = 5, col = "blue4")
```

```r
Fx <- pbinom(x, 4, 0.4)
plot(x, Fx, type='S', main = "Função de distribuição acumulada", lwd = 5, col = "magenta")
```

```r
# As probabilidades calculadas até o lançamento 4:

fx
```

```r
# Caso a moeda fosse honesta:

fx_h <- dbinom(x, 4, 0.5)
fx_h
```

Assim, percebe-se que as probabilidades sao diferentes. Isto se da porque a moeda esta viciada.

**Questão 03:** Sendo X uma variável seguindo o modelo Binomial com parâmetro n = 15 e p = 0.4, pergunta-se:

*   P(X ≥ 14)  
*   P(8 < X ≤ 10) 
*   P(X < 2 ou X ≥ 11)
*   P(X ≥ 11 ou X > 13)
*   P(X > 3 e X < 6)
*   P(X ≤ 13 | X ≥ 11)






```r
# P(X ≥ 14)

1-pbinom(13,15,0.4)
```

```r
# P(8 < X ≤ 10)

pbinom(10, 15, 0.4) - pbinom(8, 15, 0.4)
```

```r
# P(X < 2 ou X ≥ 11)

pbinom(1, 15, 0.4) + (1 - pbinom(10, 15, 0.4))
```

```r
# P(X ≥ 11 ou X > 13)

(1-pbinom(10, 15, 0.4)) + (1-pbinom(13, 15, 0.4))
```

```r
# P(X > 3 e X < 6)


c <- pbinom(6, 15, 0.4)
d <- 1-pbinom(3, 15, 0.4)

c*d
```

```r
# P(X ≤ 13 | X ≥ 11)

a<- pbinom(14, 15, 0.4) - pbinom(10, 15, 0.4)
b<- 1-pbinom(10, 15, 0.4)
a/b
```

A função de densidade descreve a verossimilhanca de uma variavel aleatoria tomar um valor dado. Ou seja, a probabilidade da variavel aleatoria cair em uma faixa particular esta sob a regiao da area do grafico.

```r
# Da funcao de densidade de uma variavel com distribuicao de Poisson com parametro lambda = 5

lambda <- 5
plot(dpois(x, lambda), type = "h", lwd = 2,
     main = "Função de distribuição de densidade",
     ylab = "P(X = x)", xlab = "Número de eventos")

legend("topright", legend = c("5"),
       title = expression(lambda), title.adj = 0.75,
       lty = 1, col = 1:3, lwd = 2, box.lty = 0)
```

```r
# Da densidade de uma variavel X ~ N(90, 100)

plot(function(x) dnorm(x, 90, 100), 60, 140, ylab='f(x)')

title('Distribuição Normal X ~ N(80, 100)')
```

```r
# Sobreponha ao grafico anterior a densidade de uma variavel Y ~ N(90, 80) e outra Z ~ N(85, 100)

N=10000 #tamanho do eixo x

y_rnorm <- rnorm(N, mean = 90, sd = 100)
y_rnorm2 <- rnorm(N, mean = 90, sd = 80)
y_rnorm3 <- rnorm(N, mean = 85, sd = 100)  

plot(density(y_rnorm),                                
     xlim = c(- 400, 600),
     ylim = c(0, 0.0055),
     main = "Densidade da distribuição normal", lwd = 2)
lines(density(y_rnorm2), col = "coral2", lwd = 2)             
lines(density(y_rnorm3), col = "green3", lwd = 2)             
legend("topright",                                    
       legend = c("Média = 100; SD = 64",
                  "Média = 90; SD = 80",
                  "Média = 85; SD = 100"),
       col = c("black", "coral2", "green3"),
       lty = 1)
```

```r
# Densidades de distribuicoes qui-quadrado com 1, 2 e 5 graus de liberdade.

library(stats)

#tamanho do eixo x
z <- 0:100
#-----------
# gl: 1
#-----------
gl <- 1
plot(dchisq(z, gl,ncp = 0, log = FALSE), type = "h", lwd = 2,
     main = "Função de distribuição de densidade",
     ylab = "P(X = x)", xlab = "Número de eventos")

#-----------
# gl: 2
#-----------
gl <- 2
lines(dchisq(z, gl,ncp = 0, log = FALSE), type = "h", lwd = 2, col = rgb(1,0,0, 0.7))

#-----------
# gl: 5
#-----------
gl <- 5
lines(dchisq(z, gl,ncp = 0, log = FALSE), type = "h", lwd = 2, col = rgb(0, 1, 0, 0.7))

# legenda
legend("topright", legend = c("1", "2", "5"),
       title = "Graus de liberdade (gl)", title.adj = 0.75,
       lty = 1, col = 1:3, lwd = 3, box.lty = 1)
```
