# 1. Medidas de Dissimilaridade: Distância Euclidiana

```r
###############################################
 ##### PRINCIPAIS CONCEITOS E MÉTRICAS #####
###############################################
```

```r
peso=c(71,50,80,52)
altura=c(1.8,1.69,1.75,1.63)
idade=c(15,16,22,25)

MAT=cbind(peso, altura, idade)
rownames(MAT)=c("Joao","Maria","Ricardo", "Luiza")
```

```r
?dist
DE=dist(MAT, method = "euclidean", diag = TRUE, upper = TRUE, p = 2)
DE
```

```r
###############################
#   Padronizacao dos dados.

MATp=apply(MAT,2,function(x) (x-mean(x))/sd(x))
DEp=dist(MATp, method = "euclidean", diag = TRUE, upper = TRUE, p = 2)
DEp
```

```r
#########################################
#  Distancia euclidianda padronizada media

DEpmedia=DEp/sqrt(nrow(MAT))
DEpmedia
```

```r
####################
# Dendrograma

dendo=hclust(DEpmedia,method = "average")
plot(dendo)
```

# 2. Medidas de Dissimilaridade: Distância de Mahalanobis

```r
D=read.table("../input/dados/Dados.txt",head=TRUE)
str(D)
D$Trat=as.factor(D$Trat)
D$Bloco=as.factor(D$Bloco)
```

```r
# Calcular a MANOVA para a construção da matriz R
#?aov()
m=aov(cbind(X1,X2,X3,X4)~Trat+Bloco,data=D)
anova(m)
```

```r
# Calcular a Soma dos Quadrados e Produtos dos Resíduos (SQP) 
SQP=crossprod(m$residuals)
SQP
```

```r
# Calcular a Matriz de (co)variância residual (CRE)
CRE=SQP/m$df.residual
CRE
```

```r
# Calcular a matriz de médias para obter a matriz de diferença das médias (V)
Med=aggregate(D[,3:6],by=list(D$Trat),mean)[,-1]
Med2=as.matrix(Med)
Med2
```

```r
# Calcular a matriz de diferença das médias (V)
V=Med2[1,]-Med2[2,]
V
```

```r
# Calcular transposta da matriz de diferença das médias (V")
t(V)%*%solve(CRE)%*%V
```

```r
# Teria que calcular para todas as combinações de diferenças
# Calcular pela função do pacote biotool
install.packages("biotools")
library(biotools)
#?biotools
```

```r
#Matriz de Distâncias de Mahalanobis
Dist=D2.dist(Med2,CRE)
Dist
```

```r
#Dendograma
dendo=hclust(Dist)
plot(dendo)
```

## 8. Dendrograma: Correlação Cofenética

```r
#library(MultivariateAnalysis)
#base de dados interna
#?Dados.MED
#data(Dados.MED)
```

```r
#Medida de dissimilaridade 
#?Distancia
# Distância euclidiana padronizada
Dist=Distancia(Dados.MED, Metodo = 4) 
Dist$Distancia
Dist

#Dendrograma
#?Dendrograma
```

```r
#Ligacao media entre grupo (UPGMA).
Dend=Dendrograma(Dist, Metodo = 3, corte = "Mojena1") 
Dend
HeatPlot(Dend)
```

```r
#A metodologia considerada para estabelecer a linha de corte no Dendrograma:

# FALSE = Indica que o ponto de corte nao sera apresentado.

# Mojena1 = Coeficiente de mojena considerando o K=1.25.

# Mojena2 = Coeficiente de mojena considerando o K=2.00.

# 'Numeric'= Valor numerico indicando onde se deseja plotar a linha de corte.
```

```r
#Avaliando as correlações cofenéticas

#?Dendrograma

#Ligacao simples (Metodo do vizinho mais proximo)
Dend1=Dendrograma(Dist, Metodo = 1, corte = "Mojena2", nperm = 9999) 
Dend1
```

```r
#Ligacao completa (Metodo do vizinho distante).
Dend2=Dendrograma(Dist, Metodo = 2, corte = "Mojena2", nperm = 9999) 
Dend2
```

```r
#Ligacao media entre grupo (UPGMA).
Dend3=Dendrograma(Dist, Metodo = 3, corte = "Mojena2", nperm = 9999) 
Dend3
```

```r
#Metodo de Ward.
Dend4=Dendrograma(Dist, Metodo = 4, corte = "Mojena2", nperm = 9999) 
Dend4
```

```r
# Metodo de Ward (d2).
Dend5=Dendrograma(Dist, Metodo = 5, corte = "Mojena2", nperm = 9999) 
Dend5
```

```r
#Metodo da mediana (WPGMC).
Dend6=Dendrograma(Dist, Metodo = 6, corte = "Mojena2", nperm = 9999) 
Dend6
```

```r
#Metodo do centroide (UPGMC).
Dend7=Dendrograma(Dist, Metodo = 7, corte = "Mojena2", nperm = 9999) 
Dend7
```

```r
#Metodo mcquitty (WPGMA
Dend8=Dendrograma(Dist, Metodo = 8, corte = "Mojena2", nperm = 9999) 
Dend8
```

**Interpretação:** 

O melhor método é o 7: Centróide, pois apresentou a melhor estimativa de correlacao cofenetica: 0.8556786

A Significancia da correlação cofenética pelo teste Mantel apresentou um pvalor: 1e-04 menor do que alfa (5%).

Portanto, rejeita Hipótese nula, ou seja, não rejeita a Hipotese Alternativa: A correlacao e maior que 0
