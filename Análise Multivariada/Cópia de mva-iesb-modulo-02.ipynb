```r
###############################################
   ##### DADOS E ANÁLISE MULTIVARIADOS #####
###############################################

# Instalar o pacote MVA
install.packages("MVA")

# Usamos dados no pacote MVA:
library("MVA")

# 'Apenas no caso de':
set.seed(280875)

# Dados multivariados surgem quando os pesquisadores registram os valores de diversas variáveis aleatórias de um
# número de sujeitos ou objetos ou talvez uma variedade de outras coisas (vamos usar o termo geral\unidades"), 
# alocando tais informações em vetores-observação multidimensionais para cada um.

# A maioria dos conjuntos de dados multivariados podem ser representados da mesma forma, ou seja, 
# em um formato retangular conhecido de planilhas, em que os elementos de cada linha correspondem aos valores das variáveis 
# de uma determinada unidade no conjunto de dados e os elementos das colunas correspondem aos valores tomados 
# por uma determinada variável.

# Embora em alguns casos onde foram coletados dados multivariados possa fazer sentido isolar cada variável e estudá-la separadamente, 
# na maioria das vezes isso não acontece. Porque todo o conjunto de variáveis é medida em cada unidade, 
# as variáveis serão relacionadas em maior ou menor grau.
# Consequentemente, se cada variável for analisada isoladamente, a estrutura completa dos dados pode não ser revelada.

# Portanto, a análise estatística multivariada é a análise estatística simultânea de uma coleção de variáveis, que melhora
# análises univariadas separadas de cada variável usando informações sobre as relações entre as variáveis.

# A análise de cada variável separadamente não permite descobrir o comportamento do fenômeno estudado
# a partir da análise conjunta das variáveis simultâneamente com o uso dos principais recursos e padrões dos dados multivariados.

### Exploratório versus Inferencial

# As unidades em um conjunto de dados multivariados às vezes são amostradas de uma população de interesse, mas geralmente não. 
# Consequentemente, existem métodos de análise multivariada que são essencialmente exploratórios e outros que podem ser usados 
# para inferência estatística.

# Exploratório: PCA, MDS, EFA e CA
# Inferencial: CFA e SEM
```

## 2. Medidas inerentes a realização das Análises Multivariadas

```r
############################################
### COVARIÂNCIAS, CORRELAÇÕES E DISTÂNCIAS
############################################

# Analisamos um conjunto de dados multivariados usando métodos multivariados em vez de olhar para cada variável separadamente 
#usando um ou outro método univariado familiar porque as estruturas ou padrões nos dados são tão prováveis de serem implícitos 
#por "relacionamentos" entre as variáveis ou pela relação " proximidade" de diferentes unidades como por seus diferentes valores 
# de variáveis. E, às vezes, por ambos.

# Portanto, podemos procurar "links" nas colunas da matriz de dados, X, 
# ou o envolvimento de subconjuntos interessantes das unidades de observação.

# Então, como quantificamos essas relações entre as variáveis e como medimos 'distâncias'?

# Precisamos das medidas de correlações, covariâncias, e distância entre os dados conjuntos das observações. 

############################################
           ### COVARIÂNCIAS ###
############################################

# A covariância de duas variáveis aleatórias é uma medida de sua dependência linear.

# A covariância populacional (teórica) de duas variáveis aleatórias, Xi e Xj, é definida
# por Cov(Xi,Xj) = E(Xi-ui)(Xj-uj) onde
# ui=E(Xi) e uj=E(Xj); E é 'expectativa'

# Se i=j a covariância de uma variável consigo mesma é sua variância.

# Se Xi e Xj são independentes uma da outra, sua covariância é necessariamente = 0

# Em um conjunto de dados multivariado com q variáveis observadas, existem q variâncias e q(q-1)/2 covariâncias


############################################
           ### CORRELAÇÕES ###
############################################

# As Covariâncias são difíceis de interpretar às vezes por causa do dimensionamento, portanto, geralmente é padronizada 
# pela divisão dos produto dos desvios padrões das duas variáveis para obter o coeficiente de correlação, pij, onde
# pij = COVij/(SDi)(SDj)

# Vantagem: É independente das escalas das duas variáveis, está entre -1 e +1 e dá uma medida da relação linear das variáveis Xi e Xj.

# No entanto, Xi e Xj podem ter uma relação não linear que não é indicada por COR

# Se tiver q variáveis, existem q(q-1)/2 correlações distintas

# A função R cor() produz uma matriz de correlação de amostra

############################################
             ### DISTÂNCIAS ###
############################################

# Para algumas técnicas multivariadas (MDS, CA) o conceito de distância entre as unidades nos dados é relevante.

# Dados os valores das variáveis para duas unidades, unidade i e unidade j, qual é uma boa medida de distância entre elas para usar?

## Distância euclidiana

# As distâncias dependem das escalas de medição, portanto, deve-se padronizá-las.

# A função scale() divide cada variável por seu desvio padrão. Em seguida, aplique a função dist().
```

## 3. Função de Densidade Normal Multivariada

```r
# Assim como a distribuição normal domina as técnicas univariadas, a distribuição normal multivariada desempenha um papel 
# importante em alguns procedimentos multivariados.

# Bom saber um pouco sobre a função densidade multivariada e como avaliar se um conjunto de dados multivariadas
# podem ser considerados como tendo essas propriedades.

# Uma propriedade de uma função de densidade normal multivariada que merece atenção é que 
# as combinações lineares das variáveis são normalmente distribuídas

# y = a1X1 + a2X2 +    + aqXq, quando 
# a1; a2; : : : ; aq é um conjunto de escalares (vetores escalares)

library("mvtnorm")

# Aqui desenhamos uma função de densidade normal bivariada com correlação p = 0,05
x <- y <- seq(from = -3, to = 3, length = 50)
x
y
dat <- as.matrix(expand.grid(x, y))
#dat

# produz densidade dado x e y

d <- dmvnorm(dat, mean = c(0, 0), 
             sigma = matrix(c(1, 0.5, 0.5, 1), 
                            ncol = 2))
d
d <- matrix(d, ncol = length(x))
#d

# desenha o gráfico da função de densidade normal multivariada
#?persp
persp(x = x, y = y, z = d, 
      xlab = "x1", 
      ylab = "x2",
      zlab = "f(x)")
```

```r
# Para muitos métodos multivariados, a suposição de normalidade multivariada não é crítica para os resultados da análise, 
# mas às vezes pode ser necessário testar a normalidade multivariada.

# Nesse sentido, pode avaliar cada variável separadamente para verificar a normalidade univariada usando um 
# gráfico de probabilidade.

# Tais gráficos são comumente aplicados em análise univariada e envolvem ordenar as observações e, em seguida, 
# plotá-las em relação aos valores apropriados de uma função de distribuição cumulativa assumida.

# Existem dois tipos básicos de gráficos para comparar duas distribuições de probabilidade.

# O gráfico de probabilidade-probabilidade e o gráfico de quantil-quantil.

# Este código desenha funções de distribuição cumulativa e quantis

x <- seq(from = -3, to = 3, length = 1000)
Fx <- pnorm(x)
Fy <- pnorm(x, mean = -1)
plot(x, Fx, type = "l", axes = FALSE, xlab = "",
     ylab = "Cumulative distribution function") 
lines(x, Fy, type = "l")
x0 <- which.min((x - 1.2)^2)
x05 <- which.min((x + 0.5)^2)
x08 <- which.min((x + 0.9)^2)
xx <- which.min(abs(Fy - Fx[x0]))
arrows(x0 = c(min(x), x[x0], x[xx], 
              x[x08], x[x08], x[x08]),
       y0 = c(Fx[x0], Fx[x0], Fy[xx], 
              0, Fx[x08], Fy[x08]), 
       x1 = c(x[x0], x[x0], x[xx], x[x08], 
              -3, -3), 
       y1 = c(Fx[x0], 0, 0, Fy[x08], 
              Fx[x08], Fy[x08]))
mtext(at = c(x[x08], x[xx], x[x0]), 
      side = 1, line = 1, text =
      c(expression(q), expression(q[2](p)), 
        expression(q[1](p))))
mtext(at = c(0, Fx[x08], Fy[x08], Fx[x0], 1), 
      line = 1, side = 2, text =
      c(0, expression(p[1](q)), 
        expression(p[2](q)), 
        expression(p), 1)) 
box()
```

```r
# Avaliar a normalidade dos dados de medição corporal

x <- measure[, c("chest", "waist", "hips")]

# # Estimar as médias de todas as três variáveis e a matriz de covariância

cm <- colMeans(x)

S <- cov(x)

# Calcular as diferenças di para todas as unidades em nossos dados, então iteramos sobre as linhas usando apply()
# Calcular a matriz das distâncias de Mahalanobis

d <- apply(x, MARGIN = 1, function(x) 
           t(x - cm) %*% solve(S) %*% (x - cm))
           
# Desenhar os gráficos de probabilidade normal de medidas de tórax, cintura e quadril separadamente:    

# Tórax
qqnorm(measure[,"chest"], main = "chest")
qqline(measure[,"chest"])

# Cintura
qqnorm(measure[,"waist"], main = "waist") 
qqline(measure[,"waist"])

# Quadril
qqnorm(measure[,"hips"], main = "hips") 
qqline(measure[,"hips"])   
```

```r
# Plotar as distâncias classificadas em relação aos quantis apropriados de distribuição qui-quadrado.

# Traçar as distâncias ordenadas em relação aos quantis correspondentes da distribuição qui-quadrada apropriada 
# deve se configurar ao longo de uma linha reta até a origem.         
plot(qchisq((1:nrow(x) - 1/2) / nrow(x), df = 3), sort(d),
     xlab = expression(paste(chi[3]^2, " Quantile")), 
     ylab = "Ordered distances")
abline(a = 0, b = 1)

# Parece se desviar um pouco da normalidade, mas temos tão poucas observações que é difícil dizer.
```

```r
# Use o gráfico qui-quadrado em dados de poluição do ar
# Plotar os gráficos de probabilidade para cada variável separada

layout(matrix(1:8, nc = 2))

# Iterar sobre todas as variáveis com sapply() que faz um loop sobre os nomes das variáveis

#install.packages("HSAUR2")
#data("USairpollution", package = "HSAUR2")
#??USairpollution

sapply(colnames(USairpollution), function(x) {
    qqnorm(USairpollution[[x]], main = x)
    qqline(USairpollution[[x]])
})
```

```r
# Os gráficos para concentração de SO2 e precipitação (precip) se desviam consideravelmente da linearidade, 
# Os gráficos para fabricação (manu) e população (popul) mostram evidências de vários valores discrepantes. 
# Mas o mais importante é o gráfico qui-quadrado para os dados.

# O script R é idêntico ao código usado para produzir o gráfico qui-quadrado para os dados de medição corporal (measure).

# Os dois pontos mais extremos do gráfico foram rotulados com os nomes das cidades às quais eles correspondem usando text().
```

```r
x <- USairpollution
cm <- colMeans(x)

S <- cov(x)

d <- apply(x, 1, function(x) t(x - cm) %*% solve(S) %*% (x - cm))
           
plot(qc <- qchisq((1:nrow(x) - 1/2) / nrow(x), df = 7), 
     sd <- sort(d),
     xlab = expression(paste(chi[7]^2, 
                             " Quantile")), 
     ylab = "Ordered distances", 
     xlim = range(qc) * c(1, 1.1))
           
# Atribuindo nome aos pontos correspondentes
           
oups <- which(rank(abs(qc - sd), 
                   ties = "random") > nrow(x) - 3)
text(qc[oups], 
     sd[oups] - 1.5, 
     names(oups))
abline(a = 0, b = 1)
```
