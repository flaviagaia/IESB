```python
def bubble_sort(lista): 
    contador = 0
    elementos = len(lista)-1 # 1 (atribuição) +1 (tamanho da lista) +1 (subtração)
    contador = contador+1
    ordenado = False # 1 (atribuição)
    contador = contador+1
    while not ordenado: # n (teste)
        contador = contador+1
        ordenado = True # 1 (atribuição)
        contador = contador+1
        for i in range(elementos): #n (teste) + 1 (cálculo range elementos)
            contador = contador+1
            if lista[i] > lista[i+1]: # 1 (comparação) +1 (indexação) +1 (indexação)
                contador = contador+1
                lista[i], lista[i+1] = lista[i+1],lista[i] # 4 (indexação) + 1 (atribuição)
                contador = contador+1
                ordenado = False # 1 (atribuição)
                contador = contador+1
                
            print(lista)
            print("Tamanho Vetor:" + str(len(lista)) + " Passos: " + str(contador))
    return lista

lista = [1,23,45,76,3,87,654,756,5345,34,534,5345,7,8779,96892,346,6,78,89]
bubble_sort(lista)

# f(n) = 5 + (n + 1) * (n + 1) * 9
# f(n) = 9n^2 + 18n + 14
# O(n^2)
```
